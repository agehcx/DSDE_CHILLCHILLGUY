```python
listToRemove = []
```

```python
for col in df.columns:
    nullCnt = df[col].isna().sum()
    print(nullCnt, end=" ")
    if nullCnt > int(0.02 * len(df[col])):
        listToRemove.append(col)
```

```python
print(len(listToRemove))
print(" ".join(listToRemove))
```

```python
df.drop(listToRemove, inplace=True, axis=1)
```
